```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatter
```

```python
def plot_random_vector_hist(n, num_samples=10000, bins=100, xlim=((-0.05, 2.05))):
    vectors = np.random.normal(0, 1/np.sqrt(n), (n, num_samples))
    norms = np.linalg.norm(vectors, axis=0)
    
    plt.figure(figsize=(7,7))
    # plt.hist(np.abs(norms - 1.), bins=bins, density=True)
    plt.hist(norms, bins=bins, density=True)
    plt.xlim(xlim)
    plt.xlabel('Norm of the random vector ($\mathit{n}$=' + str(n) + ')', fontsize=16)
    plt.ylabel('Probability density', fontsize=16)
    plt.savefig('img/Distr_' + str(n) + '.jpg')
    # plt.show()
```

```python
plot_random_vector_hist(1, num_samples=100000, xlim=None)
```

```python
plot_random_vector_hist(10, num_samples=100000)
```

```python
plot_random_vector_hist(100, num_samples=100000)
```

```python
plot_random_vector_hist(1000, num_samples=100000)
```

```python
def get_probability_epsilon(n, num_samples=10000, epsilon=0.02):
    vectors = np.random.normal(0, 1/np.sqrt(n), (n, num_samples))
    norms = np.linalg.norm(vectors, axis=0)
    
    return np.sum(np.abs(norms - 1.) > epsilon) / num_samples
```

```python
x = np.arange(100, 7000, 100, dtype=np.int32)
```

```python
y = np.array([get_probability_epsilon(xi) for xi in x])
```

```python
plt.figure(figsize=(7,7))
plt.plot(x, y)
plt.xlabel('Number of dimension', fontsize=16)
plt.ylabel('Probability ($\epsilon$=0.02)', fontsize=16)
plt.savefig('img/Proba_linear.jpg')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(x, y)
plt.yscale('log')
plt.xlabel('Number of dimension', fontsize=16)
plt.ylabel('Probability in log scale ($\epsilon$=0.02)', fontsize=16)
plt.savefig('img/Proba_log.jpg')
plt.show()
```
